## Changing Data Type

### Defined Functions
here are some functions we can use for changing the datatype

```python
def to_string(df, column_name):
    """
    Convert a pandas DataFrame column to a string.
    """
    df[column_name] = df[column_name].astype('string')
    print(f"Column Name: '{column_name}', Type: {df[column_name].dtype}")
    return df  
```

```python
def to_date(df, column_name, format=None):
    """
    Convert a pandas DataFrame column to datetime.
    """
    df[column_name] = pd.to_datetime(df[column_name], format=format)
    print(f"Column Name: '{column_name}', Type: {df[column_name].dtype}")
    return df
```

```python
def to_int(df, column_name):
    """
    Convert a pandas DataFrame column to a int.
    """
    df[column_name] = df[column_name].astype('int64')
    print(f"Column Name: '{column_name}', Type: {df[column_name].dtype}")
    return df  
```

```python
def to_float(df, column_name):
    """
    Convert a pandas DataFrame column to a float.
    """
    df[column_name] = df[column_name].astype('float64')
    print(f"Column Name: '{column_name}', Type: {df[column_name].dtype}")
    return df  
```

### Time columns
all the time columns are just integer, floats like 1210, 2123, 1020 etc which basically means 12:10, 21:23, 10:20 etc. so we add new columns which writes in date format for easier understanding

```python
def convert_hhmm(val):
    try:
        val = int(val)
        hour = val // 100
        minute = val % 100
        if 0 <= hour < 24 and 0 <= minute < 60:
            return f"{hour:02d}:{minute:02d}"
    except:
        pass
    return None
```

```python
time_columns = ['CRS_DEP_TIME', 'DEP_TIME', 'WHEELS_OFF', 'WHEELS_ON', 'CRS_ARR_TIME', 'ARR_TIME']

for col in time_columns:
    df[col.lower() + '_converted'] = df[col].apply(convert_hhmm)
    df = to_string(df, col.lower() + '_converted')
    cols = list(df.columns)
    index = cols.index(col)
    cols.remove(col.lower() + '_converted')
    cols.insert(index + 1, col.lower() + '_converted')
    df = df[cols]
```
